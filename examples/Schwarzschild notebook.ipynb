```python
import sympy as sp
from spacetime import Spacetime
```

```python
t, r, theta, phi = sp.symbols("t r theta phi", real=True)
a = sp.Function("alpha")(t,r)
b = sp.Function('beta')(t,r)
dt, dr, dtheta, dphi = sp.symbols('dt dr dtheta dphi', real=True)
dOmega2 = dtheta**2 + (sp.sin(theta)**2) * dphi**2
ds2 = sp.exp(2*a) * dt**2 - sp.exp(2*b)* dr**2 - r**2 * dOmega2
s = Spacetime((t, r, theta, phi), ds2)
```

```python
s.print_metric()
```

```python
s.print_nonzero(show_riemann=False, show_ricci=False, show_scalar=True)
```
